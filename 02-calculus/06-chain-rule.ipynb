## Concept

The chain rule is used to differentiate composite functions.

It tells us how changes in one variable affect another through intermediate functions.


## Mathematical Explanation

If:

y = f(u)  
u = g(x)

Then:

dy/dx = (dy/du) · (du/dx)

Example:

y = sin(x²)

dy/dx = cos(x²) · 2x


## Python Implementation

We use SymPy to verify chain rule derivatives.


```python
import sympy as sp

x = sp.symbols('x')
```

```python
f = sp.sin(x**2)

sp.diff(f, x)
```

## Multivariable Chain Rule

For functions with multiple variables, the chain rule extends using partial derivatives.

Used heavily in neural networks.


```python
u = x**2
v = sp.sin(u)

sp.diff(v, x)
```

## Computational Graph

Neural networks use computational graphs to apply the chain rule step by step.

Each layer computes partial derivatives and passes gradients backward.


```python
import sympy as sp

x, u = sp.symbols('x u')  # define u as a symbol
y = sp.sin(u)

# derivative dy/du
dy_du = sp.diff(y, u)

# now substitute u = x**2
dy_dx = dy_du.subs(u, x**2) * sp.diff(x**2, x)
dy_dx
```

## Visualization

Compare function and its derivative.


```python
import numpy as np
import matplotlib.pyplot as plt

f = sp.sin(x**2)
df = sp.diff(f, x)

f_l = sp.lambdify(x, f, 'numpy')
df_l = sp.lambdify(x, df, 'numpy')

x_vals = np.linspace(-3, 3, 200)

plt.plot(x_vals, f_l(x_vals), label='f(x)')
plt.plot(x_vals, df_l(x_vals), linestyle='--', label="f'(x)")

plt.legend()
plt.title("Chain Rule Example")
plt.grid()

plt.show()
```
