## Introduction

KL Divergence measures how much one probability distribution diverges from another reference distribution.

- p(x): True distribution
- q(x): Approximate / predicted distribution

It tells us how much information is lost when q is used to approximate p.

## KL Divergence Formula

D_KL(p || q) = Σ p(x) log₂ (p(x) / q(x))

Where:
- p(x): True distribution
- q(x): Predicted distribution


## Important Notes

- KL Divergence is NOT symmetric
- D_KL(p || q) ≠ D_KL(q || p)
- Always ≥ 0
- Zero only when p = q


```python
import numpy as np
import matplotlib.pyplot as plt
```

## KL Divergence Implementation

```python
def kl_divergence(p, q):
    p = np.array(p)
    q = np.array(q)
    q = np.clip(q, 1e-10, 1.0)  # avoid division by zero
    return np.sum(p * np.log2(p / q))
```

## Example

```python
# True distribution
p = [0.7, 0.2, 0.1]

# Approximate distributions
q_good = [0.6, 0.25, 0.15]
q_bad = [0.3, 0.4, 0.3]

print("KL Divergence (Good Approx):", kl_divergence(p, q_good))
print("KL Divergence (Bad Approx):", kl_divergence(p, q_bad))
```

## Visualization

```python
q = np.linspace(0.01, 0.99, 100)
p_fixed = 0.5

kl_values = [p_fixed*np.log2(p_fixed/x) + (1-p_fixed)*np.log2((1-p_fixed)/(1-x)) for x in q]

plt.plot(q, kl_values)
plt.title("KL Divergence (Binary Case)")
plt.xlabel("q (Predicted Probability)")
plt.ylabel("KL Divergence")
plt.show()
```

## Relationship with Cross-Entropy

KL Divergence = Cross-Entropy - Entropy

D_KL(p || q) = H(p, q) - H(p)
