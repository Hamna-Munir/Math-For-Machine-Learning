## Introduction

In statistics, we often want to understand how two variables are related.

- **Covariance** tells us the direction of the relationship.
- **Correlation** tells us both direction and strength.

These concepts are widely used in Machine Learning for feature analysis.

## Covariance

Covariance measures how two variables change together.

- Positive → variables increase together
- Negative → one increases while the other decreases
- Near zero → no linear relationship


```python
import numpy as np
```

## Example Data

Let's create two related variables.

```python
np.random.seed(42)
x = np.random.rand(100)
y = 2 * x + np.random.normal(0, 0.1, 100)
```

## Compute Covariance

```python
cov_matrix = np.cov(x, y)
cov_xy = cov_matrix[0, 1]

print("Covariance:", cov_xy)
```

## Correlation

Correlation is the normalized version of covariance.

- Value lies between -1 and 1
- +1 → perfect positive relationship
- -1 → perfect negative relationship
- 0 → no linear relationship


```python
corr_matrix = np.corrcoef(x, y)
corr_xy = corr_matrix[0, 1]

print("Correlation:", corr_xy)
```

## Visualization

Visualizing the relationship helps in understanding correlation.

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.title("Correlation Visualization")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## Key Differences

| Aspect | Covariance | Correlation |
|--------|-----------|------------|
| Range | (-∞, ∞) | [-1, 1] |
| Scale | Depends on units | Scale-independent |
| Interpretation | Direction only | Direction + strength |


## Why It Matters in Machine Learning

- Detect relationships between features
- Identify multicollinearity
- Feature selection and dimensionality reduction
- Used in PCA and regression models
