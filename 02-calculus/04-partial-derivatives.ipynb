## Concept

A partial derivative measures how a multivariable function changes with respect to one variable while keeping the other variables constant.

It is used when a function depends on more than one input.



## Mathematical Explanation

For a function f(x, y):

Partial derivative with respect to x:

∂f/∂x → Treat y as constant

Partial derivative with respect to y:

∂f/∂y → Treat x as constant

Example:

f(x, y) = x²y + y³

∂f/∂x = 2xy  
∂f/∂y = x² + 3y²


## Python Implementation

We use SymPy to compute partial derivatives symbolically.


```python
import sympy as sp

x, y = sp.symbols('x y')
```

```python
f = x**2 * y + y**3

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

df_dx, df_dy
```

## Gradient Vector

The gradient is a vector containing all partial derivatives.

For f(x, y):

∇f = [ ∂f/∂x , ∂f/∂y ]

It points in the direction of maximum increase.


```python
gradient = sp.Matrix([df_dx, df_dy])
gradient
```

## Visualization

We visualize the surface of the function to understand how it changes with x and y.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

f_l = sp.lambdify((x, y), f, 'numpy')

x_vals = np.linspace(-3, 3, 50)
y_vals = np.linspace(-3, 3, 50)

X, Y = np.meshgrid(x_vals, y_vals)
Z = f_l(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, alpha=0.8)
ax.set_title("Surface of f(x, y) = x²y + y³")

plt.show()
```
