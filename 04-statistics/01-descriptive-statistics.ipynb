## Introduction

Descriptive statistics helps us summarize and understand the main features of a dataset.

It provides simple numerical measures that describe the **center, spread, and shape** of data.

In this notebook, we will explore:

- Mean, Median, Mode
- Variance and Standard Deviation
- Data visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Sample Dataset

Let's create a sample dataset to compute descriptive statistics.

```python
data = np.array([10, 12, 15, 18, 20, 22, 25, 30, 35, 40])
data
```

## 1. Mean (Average)

The **mean** is the average value of the dataset.

Formula:

Mean = (Sum of all values) / (Number of values)

```python
mean = np.mean(data)
print("Mean:", mean)
```

## 2. Median

The **median** is the middle value when the data is sorted.

- If the number of values is odd → middle value
- If even → average of two middle values

```python
median = np.median(data)
print("Median:", median)
```

## 3. Mode

The **mode** is the most frequently occurring value in the dataset.

```python
mode = stats.mode(data)
print("Mode:", mode.mode[0])
```

## 4. Variance

Variance measures how far the data points are spread out from the mean.

```python
variance = np.var(data)
print("Variance:", variance)
```

## 5. Standard Deviation

Standard deviation is the square root of variance and represents spread in the same units as data.

```python
std_dev = np.std(data)
print("Standard Deviation:", std_dev)
```

## Data Visualization

Visualizing data helps us understand its distribution.

```python
plt.hist(data, bins=5)
plt.title("Data Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

## Why Descriptive Statistics Matters in Machine Learning

Descriptive statistics is the first step in any Machine Learning workflow.

It helps:

- Understand data distribution
- Detect outliers
- Prepare data for modeling
- Choose appropriate algorithms
