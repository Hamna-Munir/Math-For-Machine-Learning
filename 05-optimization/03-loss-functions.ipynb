## Introduction

A **loss function** measures how far the model's predictions are from the actual values.

The goal of training is to **minimize the loss**.

- Lower loss → Better model
- Higher loss → Poor predictions


## Types of Loss Functions

### 1. Mean Squared Error (MSE)

- Used in regression
- Penalizes large errors heavily


```python
import numpy as np

y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])

mse = np.mean((y_true - y_pred)**2)
print("MSE:", mse)
```

### 2. Mean Absolute Error (MAE)

- Used in regression
- More robust to outliers


```python
mae = np.mean(np.abs(y_true - y_pred))
print("MAE:", mae)
```

### 3. Binary Cross-Entropy

- Used in binary classification
- Based on probability outputs


```python
y_true = np.array([1, 0, 1, 0])
y_pred = np.array([0.9, 0.1, 0.8, 0.2])

epsilon = 1e-15
y_pred = np.clip(y_pred, epsilon, 1 - epsilon)

bce = -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
print("Binary Cross-Entropy:", bce)
```

## Visualization of Loss

Let's visualize how loss changes with prediction error.

```python
import matplotlib.pyplot as plt

errors = np.linspace(-5, 5, 100)
mse_loss = errors**2
mae_loss = np.abs(errors)

plt.plot(errors, mse_loss, label="MSE")
plt.plot(errors, mae_loss, label="MAE")
plt.legend()
plt.title("Loss Functions Comparison")
plt.show()
```

## Why Loss Functions Matter

- Guide optimization algorithms
- Define what "learning" means
- Affect model performance
- Different problems require different loss functions


## Choosing the Right Loss Function

- Regression → MSE, MAE
- Classification → Cross-Entropy
- Robust models → MAE or Huber Loss
