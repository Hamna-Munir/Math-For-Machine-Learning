## Introduction

In probability and statistics, we often want to understand:

- The **average value** of a random variable
- How **spread out** the values are

Two important concepts help us do this:

- **Expected Value (Mean)**
- **Variance**

These concepts are widely used in **Machine Learning, data analysis, and statistics**.

## Expected Value

The **expected value** represents the long-term average value of a random variable.

For a discrete random variable:

E(X) = Σ x * P(x)

Where:

- x = possible value
- P(x) = probability of that value


### Example: Dice Roll

Consider a fair six-sided die.

Possible values:

X = {1, 2, 3, 4, 5, 6}

Each outcome has probability 1/6.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
values = np.array([1,2,3,4,5,6])
probabilities = np.array([1/6]*6)

expected_value = np.sum(values * probabilities)

print("Expected Value:", expected_value)
```

## Variance

Variance measures **how much the values deviate from the expected value**.

Formula:

Var(X) = E[(X - μ)²]

Where:

- μ = expected value
- X = random variable

```python
variance = np.sum(((values - expected_value) ** 2) * probabilities)

print("Variance:", variance)
```

## Standard Deviation

Standard deviation is the square root of variance.

It tells us how far values typically lie from the mean.

```python
std_dev = np.sqrt(variance)

print("Standard Deviation:", std_dev)
```

## Simulation Example

Let's simulate dice rolls and estimate mean and variance.

```python
import random

rolls = 10000
data = [random.randint(1,6) for _ in range(rolls)]

print("Sample Mean:", np.mean(data))
print("Sample Variance:", np.var(data))
```

## Visualization

```python
plt.hist(data, bins=6)
plt.title("Dice Roll Distribution")
plt.xlabel("Dice Value")
plt.ylabel("Frequency")
plt.show()
```

## Why Expected Value and Variance Matter in Machine Learning

These concepts are used in many ML tasks:

- Understanding data distributions
- Feature normalization
- Gaussian models
- Loss functions
- Model evaluation

Variance also helps detect **overfitting and instability in models**.
