## Introduction

Hypothesis testing is a statistical method used to make decisions based on data.

It helps us determine whether there is enough evidence to support a claim.

In Machine Learning and Data Science, it is widely used for validating assumptions and comparing models.

## Key Concepts

- **Null Hypothesis (H₀):** Assumes no effect or no difference
- **Alternative Hypothesis (H₁):** Assumes there is an effect or difference
- **Significance Level (α):** Threshold to reject H₀ (commonly 0.05)
- **p-value:** Probability of observing results given H₀ is true


## Example Scenario

Suppose we want to test whether the mean of a dataset is equal to a specific value.

- H₀: μ = 50
- H₁: μ ≠ 50

```python
import numpy as np
from scipy import stats
```

## Generate Sample Data

We create a dataset to test our hypothesis.

```python
np.random.seed(42)
data = np.random.normal(loc=52, scale=10, size=100)
np.mean(data)
```

## Perform t-Test

We use a one-sample t-test to check whether the sample mean differs from 50.

```python
t_stat, p_value = stats.ttest_1samp(data, 50)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

## Decision Rule

- If p-value < α → Reject H₀
- If p-value ≥ α → Fail to reject H₀

Let's apply this rule.

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

## Types of Errors

- **Type I Error:** Rejecting a true null hypothesis
- **Type II Error:** Failing to reject a false null hypothesis


## Why Hypothesis Testing Matters in Machine Learning

Hypothesis testing is widely used in Machine Learning:

- Comparing models (A/B testing)
- Validating assumptions
- Feature selection
- Evaluating improvements in models
