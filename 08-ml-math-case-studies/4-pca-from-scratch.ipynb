## Introduction

Principal Component Analysis (PCA) is a **dimensionality reduction technique** used to transform data into a lower-dimensional space while preserving as much variance as possible.

Key ideas:
- Reduce number of features
- Preserve maximum variance
- Remove redundancy in data

## Steps in PCA

1. Standardize the data
2. Compute covariance matrix
3. Compute eigenvalues and eigenvectors
4. Sort eigenvectors by eigenvalues
5. Project data onto principal components

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate Sample Data

```python
np.random.seed(42)

# Create correlated data
mean = [0, 0]
cov = [[3, 2], [2, 2]]
X = np.random.multivariate_normal(mean, cov, 200)
```

## Visualize Original Data

```python
plt.scatter(X[:,0], X[:,1], alpha=0.6)
plt.title("Original Data")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

## Step 1: Standardize Data

```python
X_mean = np.mean(X, axis=0)
X_centered = X - X_mean
```

## Step 2: Covariance Matrix

```python
cov_matrix = np.cov(X_centered.T)
cov_matrix
```

## Step 3: Eigenvalues and Eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

eigenvalues, eigenvectors
```

## Step 4: Sort Eigenvectors

```python
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]
```

## Step 5: Project Data

```python
# Choose top 1 principal component
k = 1
W = eigenvectors[:, :k]

X_reduced = X_centered.dot(W)
```

## Visualize Reduced Data

```python
plt.scatter(X_reduced, np.zeros_like(X_reduced), alpha=0.6)
plt.title("1D Projection using PCA")
plt.xlabel("Principal Component 1")
plt.show()
```

## Explained Variance

```python
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)
explained_variance_ratio
```

## Conclusion

- PCA reduces dimensionality while preserving variance.
- Eigenvectors define new feature space.
- Eigenvalues show importance of each component.
- Widely used in data preprocessing, visualization, and noise reduction.
