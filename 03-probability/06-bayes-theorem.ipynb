## Introduction

Bayes' Theorem is one of the most important results in probability theory.

It allows us to **update probabilities when new information becomes available**.

Bayesian reasoning is widely used in:

- Machine Learning
- Medical diagnosis
- Spam filtering
- Recommendation systems
- Risk prediction

## Bayes' Theorem Formula

Bayes' Theorem describes the relationship between conditional probabilities.

P(A | B) = ( P(B | A) * P(A) ) / P(B)

Where:

- **P(A | B)** → Posterior probability (probability of A given B)
- **P(B | A)** → Likelihood
- **P(A)** → Prior probability
- **P(B)** → Evidence

## Intuition Behind Bayes' Theorem

Bayes' Theorem helps us answer questions like:

*If a person tests positive for a disease, what is the probability they actually have the disease?*

The theorem combines:

- Prior knowledge
- Observed evidence
- Likelihood of the evidence

## Example: Medical Test

Suppose:

- 1% of people have a disease
- The test detects the disease correctly 99% of the time
- The test gives a false positive 5% of the time

We want to compute:

P(Disease | Positive Test)

```python
P_disease = 0.01
P_no_disease = 0.99

P_positive_given_disease = 0.99
P_positive_given_no_disease = 0.05
```

## Compute Evidence P(Positive)

```python
P_positive = (P_positive_given_disease * P_disease) + (P_positive_given_no_disease * P_no_disease)

print("P(Positive Test):", P_positive)
```

## Apply Bayes' Theorem

```python
posterior = (P_positive_given_disease * P_disease) / P_positive

print("P(Disease | Positive Test):", posterior)
```

## Visualization Example

```python
import matplotlib.pyplot as plt

labels = ['Prior P(Disease)', 'Posterior P(Disease | Positive)']
values = [P_disease, posterior]

plt.bar(labels, values)
plt.title('Effect of Bayes Theorem')
plt.ylabel('Probability')
plt.show()
```
