## Introduction

In many real-world problems, computing exact integrals is difficult.

Numerical integration helps us approximate the **area under a curve** using discrete values.

This is important in:
- Probability distributions
- Physics simulations
- Machine Learning


## Basic Idea

Integral of a function represents the area under the curve.

∫ f(x) dx ≈ Sum of small areas


## Common Methods

1. Trapezoidal Rule
2. Simpson’s Rule


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define Function

```python
def f(x):
    return x**2
```

## Trapezoidal Rule

```python
def trapezoidal_rule(f, a, b, n=100):
    x = np.linspace(a, b, n)
    y = f(x)
    h = (b - a) / (n - 1)
    return (h/2) * (y[0] + 2*np.sum(y[1:-1]) + y[-1])
```

## Simpson’s Rule

```python
def simpsons_rule(f, a, b, n=100):
    if n % 2 == 1:
        n += 1
    x = np.linspace(a, b, n)
    y = f(x)
    h = (b - a) / (n - 1)
    return (h/3) * (y[0] + 4*np.sum(y[1:-1:2]) + 2*np.sum(y[2:-2:2]) + y[-1])
```

## Comparison with True Value

```python
a, b = 0, 2
true_value = (b**3)/3  # integral of x^2

trap = trapezoidal_rule(f, a, b)
simp = simpsons_rule(f, a, b)

print("True Value:", true_value)
print("Trapezoidal:", trap)
print("Simpson's:", simp)
```

## Visualization

```python
x = np.linspace(0, 2, 100)
y = f(x)

plt.plot(x, y)
plt.fill_between(x, y, alpha=0.3)
plt.title("Area under Curve (Numerical Integration)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## Effect of Number of Intervals

```python
for n in [10, 50, 100, 500]:
    approx = trapezoidal_rule(f, 0, 2, n)
    print(f"n={n}, Approx={approx}")
```

## Key Insights

- More intervals → Better accuracy
- Simpson’s Rule is generally more accurate than Trapezoidal
- Trade-off between speed and accuracy
