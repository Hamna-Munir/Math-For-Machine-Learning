## Introduction

In real-world problems, events rarely occur in isolation. Often we want to know:

- The probability of **two events happening together**.
- The probability of **one event occurring given that another event already happened**.

These ideas are captured by **Joint Probability** and **Conditional Probability**.

## Joint Probability

Joint probability measures the probability of **two events occurring together**.

It is written as:

P(A, B) or P(A ∩ B)

Example:

- Event A: It is raining
- Event B: You carry an umbrella

Joint probability represents the probability that **both A and B occur at the same time**.

## Conditional Probability

Conditional probability measures the probability of an event occurring **given that another event has already occurred**.

Formula:

P(A | B) = P(A ∩ B) / P(B)

Where:

- P(A | B) = Probability of A given B
- P(A ∩ B) = Joint probability of A and B
- P(B) = Probability of B

## Example Scenario

Suppose we survey students about:

- Whether they study Machine Learning
- Whether they know Python

We can compute joint and conditional probabilities from the data.

```python
import numpy as np
import pandas as pd
```

## Example Dataset

```python
data = {
    'ML': ['Yes','Yes','No','Yes','No','Yes','No','Yes'],
    'Python': ['Yes','Yes','Yes','No','No','Yes','Yes','Yes']
}

df = pd.DataFrame(data)
df
```

## Joint Probability Calculation

```python
joint = len(df[(df['ML']=='Yes') & (df['Python']=='Yes')]) / len(df)

print("Joint Probability P(ML and Python):", joint)
```

## Conditional Probability Calculation

```python
p_python = len(df[df['Python']=='Yes']) / len(df)

conditional = joint / p_python

print("P(ML | Python):", conditional)
```

## Independence of Events

Two events A and B are **independent** if:

P(A ∩ B) = P(A) * P(B)

If this condition holds, knowing one event does **not change** the probability of the other.

## Why Joint and Conditional Probability Matter in Machine Learning

These concepts are essential for many ML algorithms:

- Naive Bayes Classifier
- Bayesian Networks
- Probabilistic Graphical Models
- Hidden Markov Models

Understanding conditional probability is the first step toward **Bayes' Theorem**, which is widely used in Machine Learning.
