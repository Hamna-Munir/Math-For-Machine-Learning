## Introduction

Information Theory provides the mathematical foundation to measure uncertainty, information, and similarity between distributions.

In Machine Learning, it is widely used in:

- Model training (loss functions)
- Feature selection
- Probabilistic modeling
- Decision-making systems


## Key Concepts Recap

- **Entropy (H)** → Measures uncertainty
- **Cross-Entropy** → Measures prediction error
- **KL Divergence** → Measures difference between distributions
- **Mutual Information** → Measures dependency between variables


## 1. Entropy in Decision Trees

Decision Trees use **Entropy** to measure impurity in data.

Information Gain = Reduction in entropy after a split.


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris

data = load_iris()
X, y = data.data, data.target

model = DecisionTreeClassifier(criterion='entropy')
model.fit(X, y)

print("Model trained using entropy criterion")
```

## 2. Cross-Entropy as Loss Function

Cross-Entropy is widely used in classification tasks.

Lower cross-entropy → Better predictions


```python
from sklearn.metrics import log_loss

# True labels
y_true = [1, 0, 1]

# Predicted probabilities
y_pred = [0.9, 0.2, 0.8]

loss = log_loss(y_true, y_pred)
print("Cross-Entropy Loss:", loss)
```

## 3. KL Divergence in Probabilistic Models

KL Divergence is used in models like Variational Autoencoders (VAE) to compare distributions.


```python
import numpy as np

def kl_divergence(p, q):
    p = np.array(p)
    q = np.array(q)
    q = np.clip(q, 1e-10, 1.0)
    return np.sum(p * np.log2(p / q))

p = [0.6, 0.4]
q = [0.5, 0.5]

print("KL Divergence:", kl_divergence(p, q))
```

## 4. Mutual Information for Feature Selection

Mutual Information helps identify important features.


```python
from sklearn.feature_selection import mutual_info_classif

mi_scores = mutual_info_classif(X, y)
print("Feature Importance (MI):", mi_scores)
```

## Real-World Applications
