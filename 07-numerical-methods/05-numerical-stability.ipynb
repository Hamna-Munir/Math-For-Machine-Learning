## Introduction

Numerical stability is about how errors (like floating point inaccuracies) affect computation results.

- Stable algorithms → small errors remain small
- Unstable algorithms → small errors grow rapidly


## Example: Adding Small Numbers
Floating point errors can accumulate in repeated operations.

```python
import numpy as np

a = 1.0
small = 1e-16

sum_val = a
for _ in range(10**6):
    sum_val += small

print("Expected:", a + 10**6 * small)
print("Computed:", sum_val)
```

Notice how the floating point arithmetic can deviate slightly from the expected value.

## Numerical Stability in Algorithms
- **Forward stability:** algorithm produces small errors for slightly perturbed input.
- **Backward stability:** algorithm solves a slightly perturbed problem exactly.


## Example: Unstable vs Stable Matrix Solve

```python
# Unstable matrix
B = np.array([[1, 1], [1, 1.0001]])
b = np.array([2, 2.0001])

x_unstable = np.linalg.solve(B, b)
print("Solution (Unstable):", x_unstable)

# Stable approach using pseudo-inverse
x_stable = np.dot(np.linalg.pinv(B), b)
print("Solution (Stable using pinv):", x_stable)
```

Using the **pseudo-inverse** helps mitigate instability from ill-conditioned matrices.

## Tips for Numerical Stability in ML
- Normalize / scale input features.
- Avoid subtracting nearly equal numbers.
- Prefer stable algorithms (e.g., `np.linalg.pinv` over direct inversion).
- Regularization helps prevent instability in optimization.
- Use double precision if necessary (`float64`).

## Summary

- Numerical stability is crucial for accurate results.
- Floating point errors can accumulate and affect computations.
- Use stable numerical methods and regularization in ML.
- Understanding conditioning and stability helps design robust models.
