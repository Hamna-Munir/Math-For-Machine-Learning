## Introduction

In real-world problems, we usually do not have access to the entire population.

Instead, we use a sample to estimate unknown parameters such as:

- Population mean (μ)
- Population variance (σ²)

This process is called **parameter estimation**.

## Types of Estimation

There are two main types of parameter estimation:

1. Point Estimation
2. Interval Estimation

We will focus mainly on point estimation and introduce key concepts used in Machine Learning.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sample Dataset

Let's generate a dataset from a normal distribution.

```python
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(data, bins=30)
plt.title("Sample Data Distribution")
plt.show()
```

## 1. Point Estimation

Point estimation provides a single value estimate for a parameter.

Example:

- Sample mean estimates population mean
- Sample variance estimates population variance

```python
sample_mean = np.mean(data)
sample_variance = np.var(data)

print("Estimated Mean:", sample_mean)
print("Estimated Variance:", sample_variance)
```

## 2. Maximum Likelihood Estimation (MLE)

Maximum Likelihood Estimation finds parameters that maximize the probability of observing the given data.

For a normal distribution:

- Mean = sample mean
- Variance = sample variance

```python
# MLE estimates (same as sample statistics for normal distribution)
mle_mean = np.mean(data)
mle_variance = np.var(data)

print("MLE Mean:", mle_mean)
print("MLE Variance:", mle_variance)
```

## 3. Visualizing the Estimate

We can visualize how well our estimated parameters fit the data.

```python
from scipy.stats import norm

x = np.linspace(min(data), max(data), 100)
pdf = norm.pdf(x, mle_mean, np.sqrt(mle_variance))

plt.hist(data, bins=30, density=True)
plt.plot(x, pdf)
plt.title("MLE Fit to Data")
plt.show()
```

## 4. Bias in Estimation

An estimator is **biased** if it consistently overestimates or underestimates the true parameter.

Example:

- Sample variance (without correction) can be biased
- Using (n-1) in denominator gives an unbiased estimate

```python
biased_var = np.var(data)
unbiased_var = np.var(data, ddof=1)

print("Biased Variance:", biased_var)
print("Unbiased Variance:", unbiased_var)
```

## Why Parameter Estimation Matters in Machine Learning

Parameter estimation is at the core of Machine Learning:

- Models learn parameters from data
- Optimization algorithms adjust parameters
- MLE is widely used in training models
- Helps in building probabilistic models
