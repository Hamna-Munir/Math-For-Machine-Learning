## Introduction

Entropy is a fundamental concept in Information Theory that measures the **uncertainty** or **randomness** of a system.

- High entropy → More uncertainty
- Low entropy → Less uncertainty


## Entropy Formula

H(X) = - Σ p(x) log₂ p(x)

Where:
- p(x) is the probability of event x
- log₂ is logarithm base 2


## Intuition

- If all outcomes are equally likely → Maximum entropy
- If one outcome is certain → Minimum entropy (0)

Example:
- Fair coin → High entropy
- Biased coin → Lower entropy


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Entropy Function Implementation

```python
def entropy(probabilities):
    probabilities = np.array(probabilities)
    probabilities = probabilities[probabilities > 0]  # remove zeros
    return -np.sum(probabilities * np.log2(probabilities))
```

## Examples

```python
# Fair coin
fair_coin = [0.5, 0.5]
print("Entropy (Fair Coin):", entropy(fair_coin))

# Biased coin
biased_coin = [0.9, 0.1]
print("Entropy (Biased Coin):", entropy(biased_coin))
```

## Visualization

```python
p = np.linspace(0.01, 0.99, 100)
entropy_values = [- (x*np.log2(x) + (1-x)*np.log2(1-x)) for x in p]

plt.plot(p, entropy_values)
plt.title("Entropy of Bernoulli Distribution")
plt.xlabel("Probability of Success")
plt.ylabel("Entropy")
plt.show()
```

## Properties of Entropy

- Always non-negative
- Maximum when probabilities are equal
- Zero when outcome is certain


## Applications in Machine Learning

- Decision Trees (Information Gain)
- Feature selection
- Measuring uncertainty in predictions
