## Introduction

Stochastic Gradient Descent (SGD) is a variant of Gradient Descent used for faster and more efficient training.

Instead of using the entire dataset, SGD updates parameters using **one data point at a time**.


## Gradient Descent vs SGD

| Method | Data Used | Speed | Stability |
|--------|----------|-------|----------|
| Gradient Descent | Full dataset | Slow | Stable |
| SGD | One sample | Fast | Noisy |


## Why Use SGD?

- Faster for large datasets
- Uses less memory
- Can escape local minima due to randomness


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate Sample Data

```python
np.random.seed(42)
X = np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

## Initialize Parameters

```python
theta = np.random.randn(2, 1)
learning_rate = 0.1
epochs = 20
```

## Add Bias Term

```python
X_b = np.c_[np.ones((100, 1)), X]
```

## SGD Implementation

```python
for epoch in range(epochs):
    for i in range(len(X_b)):
        random_index = np.random.randint(len(X_b))
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]

        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        theta = theta - learning_rate * gradients

print("Optimized Parameters:\n", theta)
```

## Visualization

```python
plt.scatter(X, y)
plt.plot(X, X_b.dot(theta))
plt.title("SGD Linear Regression")
plt.show()
```
