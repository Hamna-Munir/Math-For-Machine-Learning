## Introduction

The **Law of Large Numbers (LLN)** states that as the number of experiments or observations increases, the **sample average approaches the expected value**.

This principle explains why probability works in practice.

For example:

- If we toss a coin many times, the proportion of heads approaches **0.5**.
- If we roll a dice many times, the average result approaches **3.5**.

## Mathematical Idea

If X₁, X₂, X₃, ..., Xₙ are independent random variables with the same expected value μ, then:

Sample Mean → Expected Value as n → ∞

This means the average of many observations becomes very close to the true mean.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Coin Toss Simulation

Let's simulate many coin tosses and observe how the average changes over time.

```python
n = 1000

coin_tosses = np.random.choice([0,1], size=n)

running_mean = np.cumsum(coin_tosses) / np.arange(1, n+1)
```

## Visualization

The graph below shows how the sample mean approaches the expected value.

```python
plt.plot(running_mean)
plt.axhline(0.5)
plt.title("Law of Large Numbers - Coin Toss")
plt.xlabel("Number of Tosses")
plt.ylabel("Sample Mean")
plt.show()
```

## Dice Simulation Example

Let's repeat the same idea using a dice roll.

```python
n = 1000

dice_rolls = np.random.randint(1,7,n)

running_mean_dice = np.cumsum(dice_rolls) / np.arange(1,n+1)
```

```python
plt.plot(running_mean_dice)
plt.axhline(3.5)
plt.title("Law of Large Numbers - Dice Roll")
plt.xlabel("Number of Rolls")
plt.ylabel("Sample Mean")
plt.show()
```

## Why the Law of Large Numbers Matters in Machine Learning

The Law of Large Numbers explains why **large datasets produce reliable models**.

Applications in ML:

- Model training improves with more data
- Reliable estimation of probabilities
- Stable parameter estimation
- Reduced randomness in predictions
