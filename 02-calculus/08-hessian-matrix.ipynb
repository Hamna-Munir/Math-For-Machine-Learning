## Introduction

The Hessian matrix contains all second-order partial derivatives of a function.

It describes the curvature of multivariable functions.


## Definition

For a function f(x, y):

The Hessian matrix H is:

H =
[ ∂²f/∂x²    ∂²f/∂x∂y ]  
[ ∂²f/∂y∂x   ∂²f/∂y² ]


## Mathematical Example

Let:

f(x, y) = x² + y² + xy

Then:

∂²f/∂x² = 2  
∂²f/∂y² = 2  
∂²f/∂x∂y = 1


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = sp.symbols('x y')

f = x**2 + y**2 + x*y

f
```

## Second-Order Partial Derivatives

We compute second derivatives using SymPy.


```python
f_xx = sp.diff(f, x, 2)
f_yy = sp.diff(f, y, 2)
f_xy = sp.diff(f, x, y)
f_yx = sp.diff(f, y, x)

f_xx, f_yy, f_xy, f_yx
```

## Constructing the Hessian Matrix


```python
H = sp.Matrix([
    [f_xx, f_xy],
    [f_yx, f_yy]
])

H
```

## Second Derivative Test

The Hessian helps classify critical points:

- Positive definite → Minimum  
- Negative definite → Maximum  
- Indefinite → Saddle point


```python
H.eigenvals()
```

## Visualization of Curvature


```python
f_l = sp.lambdify((x, y), f, 'numpy')

x_vals = np.linspace(-4, 4, 100)
y_vals = np.linspace(-4, 4, 100)

X, Y = np.meshgrid(x_vals, y_vals)
Z = f_l(X, Y)

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, alpha=0.7)

ax.set_title("Surface: f(x, y) = x² + y² + xy")

plt.show()
```
