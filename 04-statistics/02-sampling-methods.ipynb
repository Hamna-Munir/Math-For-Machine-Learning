## Introduction

In real-world scenarios, it is often impossible to collect data from an entire population.

Instead, we use **sampling techniques** to select a subset of data that represents the whole population.

Good sampling ensures accurate and reliable statistical analysis.

## Types of Sampling Methods

Sampling methods are broadly divided into:

- Random Sampling
- Systematic Sampling
- Stratified Sampling
- Cluster Sampling

We will explore each method with examples.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Population Dataset

Let's create a synthetic population dataset.

```python
population = np.arange(1, 1001)  # Population of 1000 values
population[:10]
```

## 1. Random Sampling

Random sampling selects data points randomly from the population.

Each data point has an equal chance of being selected.

```python
random_sample = np.random.choice(population, size=100, replace=False)
random_sample[:10]
```

## 2. Systematic Sampling

Systematic sampling selects every k-th element from the population.

```python
k = 10
systematic_sample = population[::k]
systematic_sample[:10]
```

## 3. Stratified Sampling

Stratified sampling divides the population into groups (strata) and samples from each group.

```python
group1 = population[:500]
group2 = population[500:]

sample1 = np.random.choice(group1, 50, replace=False)
sample2 = np.random.choice(group2, 50, replace=False)

stratified_sample = np.concatenate([sample1, sample2])
stratified_sample[:10]
```

## 4. Cluster Sampling

Cluster sampling divides the population into clusters and randomly selects entire clusters.

```python
clusters = np.array_split(population, 10)

selected_cluster = clusters[np.random.randint(0, 10)]
selected_cluster[:10]
```

## Visualization of Samples

Let's compare the population and one sample.

```python
plt.hist(population, bins=30)
plt.title("Population Distribution")
plt.show()

plt.hist(random_sample, bins=30)
plt.title("Random Sample Distribution")
plt.show()
```

## Why Sampling Matters in Machine Learning

Sampling plays a crucial role in Machine Learning:

- Training models on subsets of data
- Reducing computational cost
- Handling large datasets
- Preventing bias in model training
- Creating train-test splits
