## Introduction

Optimization is the core of Machine Learning.

Every ML model learns by minimizing a **loss function** using optimization algorithms.


## Optimization Pipeline in ML

1. Define a model
2. Choose a loss function
3. Compute gradients
4. Update parameters
5. Repeat until convergence


## Example: Linear Regression with Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate Data

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

## Add Bias Term

```python
X_b = np.c_[np.ones((100, 1)), X]
```

## Initialize Parameters

```python
theta = np.random.randn(2, 1)
learning_rate = 0.1
iterations = 50
```

## Gradient Descent Optimization

```python
m = len(X_b)

for i in range(iterations):
    gradients = (2/m) * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients

print("Optimized Parameters:\n", theta)
```

## Predictions

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_pred = X_new_b.dot(theta)

print("Predictions:\n", y_pred)
```

## Visualization

```python
plt.scatter(X, y)
plt.plot(X_new, y_pred, color='red')
plt.title("Linear Regression using Gradient Descent")
plt.show()
```

## Where Optimization is Used in ML

- Linear Regression → Gradient Descent
- Logistic Regression → Optimization of log-loss
- Neural Networks → Backpropagation + optimizers
- Deep Learning → Adam, RMSProp, SGD


## Challenges in Optimization

- Local minima
- Saddle points
- Vanishing gradients
- Learning rate tuning


## Best Practices

- Normalize data
- Use adaptive optimizers (Adam)
- Tune learning rate
- Use regularization
