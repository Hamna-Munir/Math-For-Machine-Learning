## Introduction

Taylor Series approximates complex functions using polynomials around a point.

It allows us to simplify nonlinear functions for analysis and optimization.


## Definition

For a function f(x), the Taylor Series around x = a is:

f(x) = f(a) + f'(a)(x−a) + f''(a)/2!(x−a)² + ...

If a = 0, it is called the Maclaurin Series.


## Mathematical Example

Taylor Series of eˣ around 0:

eˣ = 1 + x + x²/2! + x³/3! + ...


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')

f = sp.exp(x)

sp.series(f, x, 0, 6)
```

## Polynomial Approximation

We approximate functions using low-degree polynomials.


```python
taylor_poly = sp.series(f, x, 0, 4).removeO()

taylor_poly
```

## Numerical Evaluation

Compare true function and Taylor approximation.


```python
f_l = sp.lambdify(x, f, 'numpy')
t_l = sp.lambdify(x, taylor_poly, 'numpy')

x_vals = np.linspace(-2, 2, 200)

plt.plot(x_vals, f_l(x_vals), label='e^x')
plt.plot(x_vals, t_l(x_vals), linestyle='--', label='Taylor Approx')

plt.legend()
plt.title("Taylor Approximation of e^x")
plt.grid()

plt.show()
```

## Second-Order Approximation

Second-order Taylor expansion is used in optimization.


```python
sp.series(f, x, 0, 3)
```

## Machine Learning Connection

Taylor Series is used in:

- Newton’s Method
- Loss approximation
- Optimization algorithms
- Convergence analysis
- Hessian-based learning
