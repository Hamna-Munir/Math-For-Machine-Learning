## Concept
The rank of a matrix is the maximum number of linearly independent rows or columns. It represents the amount of unique information contained in the matrix.

## Mathematical Explanation
For a matrix A, the rank is the dimension of its column space and row space. It determines whether a system of linear equations has a unique solution, infinitely many solutions, or no solution.

## Python Code
The rank of a matrix can be computed using NumPy's linear algebra module.

```python
import numpy as np

A = np.array([[1, 2, 3],
              [2, 4, 6],
              [1, 1, 1]])

np.linalg.matrix_rank(A)
```

## Visualization
The heatmap below shows the structure of the matrix whose rank is being analyzed.

```python
import matplotlib.pyplot as plt

plt.imshow(A, cmap='viridis')
plt.colorbar()
plt.title('Matrix Heatmap')
plt.show()
```

## ML Connection
Matrix rank determines the effective dimensionality of data. Low-rank matrices appear in dimensionality reduction, recommendation systems, and compressed representations.

## Summary
- Rank measures the number of independent rows or columns
- It indicates data redundancy
- Rank is essential for solving linear systems and ML optimization
