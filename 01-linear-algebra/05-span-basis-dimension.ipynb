## Concept
Span describes all possible linear combinations of a set of vectors. A basis is the smallest set of linearly independent vectors that spans a vector space. The dimension of a vector space is the number of vectors in its basis.

## Mathematical Explanation
If vectors v₁, v₂, ..., vₙ span a space V, then any vector in V can be written as a linear combination of these vectors. A basis is a set of vectors that both spans the space and is linearly independent. The dimension of the space equals the number of basis vectors.

## Python Code
The following example checks linear independence and determines the basis using matrix rank.

```python
import numpy as np

vectors = np.array([[1, 0], [0, 1], [1, 1]])
rank = np.linalg.matrix_rank(vectors)

rank
```

## Visualization
The plot illustrates how basis vectors span a two-dimensional space.

```python
import matplotlib.pyplot as plt

v1 = np.array([1, 0])
v2 = np.array([0, 1])

plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, label='v2')

plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.grid()
plt.legend()
plt.title('Basis Vectors Spanning 2D Space')
plt.show()
```

## ML Connection
In Machine Learning, basis vectors represent independent features. Dimensionality reduction techniques such as PCA aim to find a lower-dimensional basis that preserves most of the data variance.

## Summary
- Span defines all possible linear combinations of vectors
- A basis is the minimal spanning and independent set
- Dimension measures the complexity of the feature space
