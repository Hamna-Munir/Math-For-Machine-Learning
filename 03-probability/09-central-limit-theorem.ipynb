## Introduction

The **Central Limit Theorem (CLT)** is one of the most important concepts in statistics.

It states that **the distribution of the sample mean approaches a normal distribution as the sample size increases**, regardless of the original distribution of the data.

This is extremely powerful because it allows us to apply normal distribution methods to many real-world problems.

## Key Idea

If we take many samples from any distribution and compute their averages:

- The distribution of those averages will become **approximately normal**.
- This happens when the sample size is sufficiently large (usually n ≥ 30).

This property is what allows statisticians and ML engineers to make reliable predictions.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Example: Uniform Distribution

We will start with a **uniform distribution** (not normal).

Then we will compute the means of many samples to see how the distribution changes.

```python
data = np.random.uniform(0, 10, 10000)

plt.hist(data, bins=30)
plt.title("Original Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Sampling Process

Now we will:

- Take many random samples
- Compute the mean of each sample
- Observe the distribution of those means

```python
sample_means = []

for _ in range(1000):
    sample = np.random.choice(data, size=30)
    sample_means.append(np.mean(sample))
```

## Distribution of Sample Means

According to the Central Limit Theorem, the distribution of sample means should look **approximately normal**.

```python
plt.hist(sample_means, bins=30)
plt.title("Distribution of Sample Means (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

## Why CLT Matters in Machine Learning

The Central Limit Theorem is important because many ML techniques rely on it.

Applications include:

- Statistical inference
- Confidence intervals
- Hypothesis testing
- Estimating model performance
- Understanding noise in data

## Real-World Importance

Even if real-world data is not normally distributed, the CLT allows us to work with **normal approximations** when dealing with averages.

This makes statistical analysis and Machine Learning much easier.
