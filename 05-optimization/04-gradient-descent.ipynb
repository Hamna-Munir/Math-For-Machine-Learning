## Introduction

Gradient Descent is one of the most important optimization algorithms in Machine Learning.

It helps us find the **minimum of a function** by iteratively moving in the direction of the negative gradient.

- Start with an initial value
- Move step by step toward the minimum
- Update parameters using gradients


## Intuition

Imagine standing on a hill:

- You look at the slope
- Move downward step by step
- Eventually reach the lowest point

This is exactly how Gradient Descent works.

## Update Rule

Gradient Descent updates parameters using:

x = x - α * gradient

Where:

- α (alpha) → Learning rate
- Gradient → Slope of the function


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Example Function

We use a simple function:

f(x) = x²

Its gradient is:

f'(x) = 2x

```python
def f(x):
    return x**2

def grad_f(x):
    return 2*x
```

## Gradient Descent Implementation

```python
x = 8  # initial value
learning_rate = 0.1
iterations = 20

history = []

for i in range(iterations):
    grad = grad_f(x)
    x = x - learning_rate * grad
    history.append(x)

print("Minimum found at x =", x)
```

## Visualization

```python
x_vals = np.linspace(-10, 10, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals)
plt.scatter(history, [f(i) for i in history])
plt.title("Gradient Descent Optimization")
plt.show()
```

## Learning Rate

- Too small → Slow convergence
- Too large → May overshoot or diverge

Choosing the right learning rate is crucial.

## Why Gradient Descent Matters

- Used in almost all ML algorithms
- Backbone of Neural Networks
- Helps minimize loss functions
