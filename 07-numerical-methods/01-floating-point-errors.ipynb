## Introduction

Computers cannot store real numbers with perfect precision. Instead, they use **floating-point representation**, which can introduce small errors.

These errors can accumulate and affect Machine Learning models and numerical computations.

## Example of Floating Point Error


```python
0.1 + 0.2
```

Expected result: 0.3  
Actual result may be: 0.30000000000000004

This happens due to binary representation limitations.

## Why This Happens

Floating-point numbers are stored in binary format, and some decimal numbers cannot be represented exactly.

Similar to how 1/3 cannot be exactly represented in decimal.

```python
format(0.1, '.20f')
```

## Accumulation of Errors


```python
sum_val = 0.0
for _ in range(10):
    sum_val += 0.1

sum_val
```

Even small errors can accumulate over multiple operations.

## Comparing Floating Point Numbers

Avoid direct comparison using ==


```python
(0.1 + 0.2) == 0.3
```

## Correct Way: Use Tolerance

```python
import math

math.isclose(0.1 + 0.2, 0.3)
```

## Visualization of Error Growth

```python
import matplotlib.pyplot as plt

values = []
errors = []

sum_val = 0.0
for i in range(1, 101):
    sum_val += 0.1
    values.append(sum_val)
    errors.append(abs(sum_val - i*0.1))

plt.plot(errors)
plt.title("Floating Point Error Growth")
plt.xlabel("Iterations")
plt.ylabel("Error")
plt.show()
```

## Best Practices

- Avoid equality checks with floats
- Use tolerance-based comparisons
- Be careful with repeated additions
- Use libraries like NumPy for stable computations
