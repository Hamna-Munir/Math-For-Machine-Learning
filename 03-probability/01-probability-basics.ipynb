## Introduction

Probability measures the likelihood that an event will occur.

In Machine Learning, probability helps models deal with **uncertainty, noise, and predictions**.

Examples:

- Spam detection
- Medical diagnosis
- Recommendation systems
- Risk prediction


## Key Concepts

**Experiment** – A process that produces outcomes.

**Outcome** – A possible result of an experiment.

**Sample Space (S)** – Set of all possible outcomes.

**Event** – A subset of the sample space.


### Example

Tossing a coin:

Sample Space:

S = {Heads, Tails}

Event:

A = Getting Heads


## Probability Formula

If all outcomes are equally likely:

P(A) = Number of favorable outcomes / Total outcomes

Properties:

- 0 ≤ P(A) ≤ 1
- P(S) = 1
- P(∅) = 0


```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Coin Toss Simulation

Let's simulate a coin toss experiment using Python.

```python
tosses = 1000

results = [random.choice(['H', 'T']) for _ in range(tosses)]

heads = results.count('H')
tails = results.count('T')

print('Heads:', heads)
print('Tails:', tails)
```

## Empirical Probability

Empirical probability is calculated from experimental results.

P(A) = Number of times event A occurs / Total trials


```python
p_heads = heads / tosses
p_tails = tails / tosses

print('P(Heads) ≈', p_heads)
print('P(Tails) ≈', p_tails)
```

## Visualization

Visualizing the probability results.

```python
labels = ['Heads', 'Tails']
values = [heads, tails]

plt.bar(labels, values)
plt.title('Coin Toss Outcomes')
plt.ylabel('Count')
plt.show()
```

## Why Probability Matters in Machine Learning

Probability is the foundation of many ML algorithms.

Understanding probability helps us build models that can reason under uncertainty.
