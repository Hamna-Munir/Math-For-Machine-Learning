## Introduction

A **random variable** is a variable whose value is determined by the outcome of a random experiment.

Instead of directly working with outcomes like *Heads or Tails*, we assign **numerical values** to outcomes.

Random variables are fundamental in statistics, probability theory, and Machine Learning.

## Example

Consider tossing a coin.

We can define a random variable X:

- X = 1 if Heads occurs
- X = 0 if Tails occurs

This converts outcomes into numbers which can be analyzed mathematically.

## Types of Random Variables

### 1. Discrete Random Variables

A discrete random variable takes **countable values**.

Examples:

- Number of heads in coin tosses
- Number of emails received in a day
- Number of customers visiting a store

---

### 2. Continuous Random Variables

A continuous random variable can take **any value in a range**.

Examples:

- Height of a person
- Temperature
- Time required to train a model

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Discrete Random Variable Example

Let's simulate rolling a dice.

Possible values:

X ∈ {1,2,3,4,5,6}

```python
rolls = 1000

dice_results = [random.randint(1,6) for _ in range(rolls)]

print(dice_results[:10])
```

## Frequency Distribution

We can observe how frequently each outcome occurs.

```python
values, counts = np.unique(dice_results, return_counts=True)

for v, c in zip(values, counts):
    print(f"Value {v}: {c} times")
```

## Visualization

Plotting the distribution of dice outcomes.

```python
plt.bar(values, counts)
plt.xlabel('Dice Value')
plt.ylabel('Frequency')
plt.title('Distribution of Dice Rolls')
plt.show()
```

## Continuous Random Variable Example

Let's generate random values from a continuous distribution.

```python
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30)
plt.title('Normal Distribution Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Why Random Variables Matter in Machine Learning

Random variables are used throughout Machine Learning:

- Modeling uncertainty
- Probability distributions
- Bayesian inference
- Feature distributions
- Stochastic optimization

Many ML models assume that data follows specific probability distributions.
