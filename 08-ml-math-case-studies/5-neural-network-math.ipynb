## Overview

A Neural Network consists of:
- Input Layer
- Hidden Layer(s)
- Output Layer

Core components:
- Weights (W)
- Biases (b)
- Activation functions
- Loss function

## Mathematical Representation

Forward propagation:

Z = W·X + b  
A = activation(Z)

We stack layers to build deep networks.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Activation Functions

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)

def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float)
```

## Simple Neural Network (1 Hidden Layer)

Architecture:
- Input → Hidden → Output

```python
np.random.seed(42)

# Sample dataset (XOR-like)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])
```

## Initialize Parameters

```python
input_size = 2
hidden_size = 4
output_size = 1

W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))

W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))
```

## Forward Propagation

```python
def forward(X):
    Z1 = X.dot(W1) + b1
    A1 = relu(Z1)

    Z2 = A1.dot(W2) + b2
    A2 = sigmoid(Z2)

    return Z1, A1, Z2, A2
```

## Loss Function (Binary Cross-Entropy)

```python
def compute_loss(y, y_hat):
    m = y.shape[0]
    loss = -(1/m) * np.sum(y*np.log(y_hat + 1e-8) + (1-y)*np.log(1-y_hat + 1e-8))
    return loss
```

## Backpropagation

```python
def backward(X, y, Z1, A1, Z2, A2):
    m = X.shape[0]

    dZ2 = A2 - y
    dW2 = (A1.T.dot(dZ2)) / m
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m

    dA1 = dZ2.dot(W2.T)
    dZ1 = dA1 * relu_derivative(Z1)
    dW1 = (X.T.dot(dZ1)) / m
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m

    return dW1, db1, dW2, db2
```

## Training Loop

```python
learning_rate = 0.1
epochs = 5000

losses = []

for i in range(epochs):
    Z1, A1, Z2, A2 = forward(X)
    loss = compute_loss(y, A2)
    losses.append(loss)

    dW1, db1, dW2, db2 = backward(X, y, Z1, A1, Z2, A2)

    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    if i % 500 == 0:
        print(f"Epoch {i}, Loss: {loss:.4f}")
```

## Loss Curve

```python
plt.plot(losses)
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

## Predictions

```python
_, _, _, predictions = forward(X)
print("Predictions:\n", predictions.round())
```

## Conclusion

- Neural networks learn by adjusting weights using gradients.
- Forward pass computes predictions.
- Backpropagation computes gradients.
- Gradient descent updates parameters.

This is the mathematical foundation of deep learning.
