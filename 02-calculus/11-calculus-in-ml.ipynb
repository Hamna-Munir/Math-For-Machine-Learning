## Loss Functions

A loss function measures how far predictions are from actual values.

Example:

Mean Squared Error (MSE):

L = (1/n) Σ (y − ŷ)²


## Gradients in Optimization

The gradient shows the direction of steepest increase.

In ML, we move in the opposite direction to minimize loss.


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Mean Squared Error
def mse(y, y_pred):
    return np.mean((y - y_pred)**2)

y = np.array([2, 4, 6, 8])
y_pred = np.array([2.2, 3.8, 5.9, 8.1])

mse(y, y_pred)
```

## Gradient Descent

Gradient Descent is an optimization algorithm.

Update rule:

θ = θ − α ∇L(θ)

Where:
- θ = parameters
- α = learning rate
- ∇L = gradient


```python
# Minimize f(x) = x^2 using Gradient Descent

def f(x):
    return x**2

def grad_f(x):
    return 2*x

x = 10
lr = 0.1

history = []

for i in range(20):
    x = x - lr * grad_f(x)
    history.append(x)

history
```

## Optimization Visualization


```python
x_vals = np.linspace(-10, 10, 200)
y_vals = x_vals**2

plt.plot(x_vals, y_vals)

for i in history:
    plt.scatter(i, f(i))

plt.title("Gradient Descent Optimization")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.show()
```

## Partial Derivatives in ML

In multivariable models, we use partial derivatives.

Each parameter has its own gradient.


```python
x, y = sp.symbols('x y')

f = x**2 + y**2

sp.diff(f, x), sp.diff(f, y)
```
