## Introduction

A **probability distribution** describes how probabilities are assigned to the values of a random variable.

It tells us **how likely each possible outcome is**.

Probability distributions are widely used in **statistics, data science, and machine learning** to model uncertainty and real-world data.

## Types of Probability Distributions

There are two main types:

### 1. Discrete Probability Distribution

Used when the random variable takes **countable values**.

Examples:
- Dice rolls
- Number of emails received
- Number of students in a class

Discrete distributions use a **Probability Mass Function (PMF)**.

### 2. Continuous Probability Distribution

Used when the random variable can take **any value within a range**.

Examples:
- Height
- Temperature
- Time

Continuous distributions use a **Probability Density Function (PDF)**.

## Discrete Distribution Example (Dice Roll)

Let's simulate dice rolls and observe the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
rolls = 1000

dice_results = [random.randint(1,6) for _ in range(rolls)]
```

### Frequency of Outcomes

```python
values, counts = np.unique(dice_results, return_counts=True)

for v, c in zip(values, counts):
    print(f"Value {v}: {c} times")
```

### Visualization of Discrete Distribution

```python
plt.bar(values, counts)
plt.xlabel('Dice Value')
plt.ylabel('Frequency')
plt.title('Discrete Probability Distribution - Dice Roll')
plt.show()
```

## Continuous Distribution Example

Now let's generate data from a **normal distribution**, one of the most important distributions in Machine Learning.

```python
data = np.random.normal(loc=0, scale=1, size=1000)
```

### Visualization of Normal Distribution

```python
plt.hist(data, bins=30)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Important Probability Functions

### Probability Mass Function (PMF)

Used for **discrete random variables**.

It gives the probability that a random variable takes a specific value.

---

### Probability Density Function (PDF)

Used for **continuous random variables**.

Instead of exact probabilities, it describes the **density of probability**.

---

### Cumulative Distribution Function (CDF)

The probability that a random variable is **less than or equal to a certain value**.

## Why Probability Distributions Matter in Machine Learning

Probability distributions help us:

- Model real-world data
- Estimate likelihoods
- Understand uncertainty
- Build probabilistic models

Many Machine Learning algorithms rely on probability distributions, including:

- Naive Bayes
- Gaussian Mixture Models
- Hidden Markov Models
- Bayesian Learning
