## Introduction

Mutual Information (MI) measures how much information one random variable contains about another.

- High MI → Strong dependency
- Low MI → Weak or no dependency

It helps answer: *How much does knowing X reduce uncertainty about Y?*

## Mutual Information Formula

I(X; Y) = Σ p(x, y) log₂ ( p(x, y) / (p(x)p(y)) )

Alternative form:

I(X; Y) = H(X) + H(Y) - H(X, Y)


## Key Properties

- Always ≥ 0
- Symmetric: I(X;Y) = I(Y;X)
- Zero when variables are independent


```python
import numpy as np
from sklearn.metrics import mutual_info_score
```

## Mutual Information Implementation (Using sklearn)

```python
# Example discrete variables
X = [0, 0, 1, 1]
Y = [0, 1, 0, 1]

mi = mutual_info_score(X, Y)
print("Mutual Information:", mi)
```

## Manual Implementation (Conceptual)

```python
def mutual_information(p_xy, p_x, p_y):
    mi = 0
    for i in range(len(p_xy)):
        if p_xy[i] > 0:
            mi += p_xy[i] * np.log2(p_xy[i] / (p_x[i] * p_y[i]))
    return mi
```

## Intuition

- If X and Y are independent → MI = 0
- If knowing X perfectly predicts Y → High MI

Example:
- Height & Weight → High MI
- Dice roll & Weather → Low MI


## Visualization Idea

Try comparing mutual information between:

- Random variables (low MI)
- Correlated variables (high MI)


```python
# Random variables
np.random.seed(0)
X = np.random.randint(0, 2, 100)
Y = np.random.randint(0, 2, 100)

print("MI (Independent):", mutual_info_score(X, Y))

# Correlated variables
Y = X.copy()
print("MI (Perfectly Correlated):", mutual_info_score(X, Y))
```
