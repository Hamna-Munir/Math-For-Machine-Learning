## Introduction

Integration is the reverse process of differentiation.

It is used to compute areas, probabilities, and expectations in Machine Learning.


## Definition

An integral represents accumulation.

Indefinite Integral:
∫ f(x) dx = F(x) + C

Definite Integral:
∫[a,b] f(x) dx = Area under the curve


## Fundamental Theorem of Calculus

If F'(x) = f(x), then:

∫[a,b] f(x) dx = F(b) − F(a)


## Applications of Integration

- Probability density functions  
- Expected values  
- Continuous distributions  
- Normalization constants  
- Bayesian inference


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')

f = x**2

sp.integrate(f, x)
```

## Definite Integral Example

Compute area under x² from 0 to 2.


```python
sp.integrate(f, (x, 0, 2))
```

## Numerical Integration

When symbolic integration is difficult, numerical methods are used.


```python
from scipy import integrate

f_num = lambda x: x**2

integrate.quad(f_num, 0, 2)
```

## Visualization of Integral


```python
x_vals = np.linspace(0, 2, 200)
y_vals = x_vals**2

plt.plot(x_vals, y_vals)
plt.fill_between(x_vals, y_vals, alpha=0.3)

plt.title("Area Under f(x) = x² from 0 to 2")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.show()
```

## Probability Connection

For a probability density function p(x):

∫ p(x) dx = 1

This ensures total probability is 1.
