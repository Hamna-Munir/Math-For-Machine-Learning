## Introduction

In Machine Learning, models can sometimes perform very well on training data but poorly on unseen data.

This problem is called **overfitting**.

Regularization helps reduce overfitting by adding a penalty to the model.


## What is Regularization?

Regularization adds a penalty term to the loss function:

New Loss = Original Loss + Penalty


## Types of Regularization

1. L1 Regularization (Lasso)
2. L2 Regularization (Ridge)
3. Elastic Net


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate Sample Data

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

## Add Polynomial Features (to cause overfitting)

```python
X_poly = np.c_[X, X**2, X**3, X**4, X**5]
```

## Normal Equation (Without Regularization)

```python
X_b = np.c_[np.ones((100, 1)), X_poly]
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

## L2 Regularization (Ridge)

```python
lambda_val = 1
I = np.eye(X_b.shape[1])
I[0, 0] = 0  # Do not regularize bias term

theta_ridge = np.linalg.inv(X_b.T.dot(X_b) + lambda_val * I).dot(X_b.T).dot(y)
```

## L1 Regularization (Concept)

L1 adds absolute values of weights:

Penalty = λ * Σ|w|

It encourages sparsity (many weights become zero).

## Visualization

```python
X_plot = np.linspace(0, 2, 100).reshape(100, 1)
X_plot_poly = np.c_[X_plot, X_plot**2, X_plot**3, X_plot**4, X_plot**5]
X_plot_b = np.c_[np.ones((100, 1)), X_plot_poly]

y_pred = X_plot_b.dot(theta)
y_ridge_pred = X_plot_b.dot(theta_ridge)

plt.scatter(X, y)
plt.plot(X_plot, y_pred, label='Without Regularization')
plt.plot(X_plot, y_ridge_pred, label='Ridge Regularization')

plt.legend()
plt.title("Effect of Regularization")
plt.show()
```

## Why Regularization Matters

- Prevents overfitting
- Improves generalization
- Controls model complexity
