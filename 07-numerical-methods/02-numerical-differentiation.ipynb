## Introduction

In many real-world problems, computing the exact derivative is difficult or impossible.

Numerical differentiation helps us approximate derivatives using function values.

This is especially important in:
- Optimization
- Gradient Descent
- Deep Learning


## Basic Idea

Derivative of a function f(x) is:

f'(x) ≈ (f(x + h) - f(x)) / h

Where h is a small number.

## Types of Numerical Differentiation

1. Forward Difference
2. Backward Difference
3. Central Difference (more accurate)


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define Function

```python
def f(x):
    return x**2  # simple function
```

## Numerical Derivatives

```python
def forward_diff(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h

def backward_diff(f, x, h=1e-5):
    return (f(x) - f(x - h)) / h

def central_diff(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2*h)
```

## Comparison with True Derivative

```python
x = 2
true_value = 2 * x  # derivative of x^2

print("True Derivative:", true_value)
print("Forward Diff:", forward_diff(f, x))
print("Backward Diff:", backward_diff(f, x))
print("Central Diff:", central_diff(f, x))
```

## Visualization

```python
x_vals = np.linspace(-5, 5, 100)
true_derivative = 2 * x_vals

approx_derivative = [central_diff(f, x) for x in x_vals]

plt.plot(x_vals, true_derivative, label="True Derivative")
plt.plot(x_vals, approx_derivative, linestyle='--', label="Numerical (Central)")
plt.legend()
plt.title("Numerical vs True Derivative")
plt.show()
```

## Effect of Step Size (h)


```python
h_values = [1e-1, 1e-3, 1e-5, 1e-7]

for h in h_values:
    approx = central_diff(f, 2, h)
    print(f"h={h}, Approx Derivative={approx}")
```

## Key Insights

- Smaller h → More accurate (up to a limit)
- Too small h → Floating point errors
- Central difference is more accurate than forward/backward
