## Concept
Singular Value Decomposition (SVD) factorizes any matrix into three components that reveal its geometric and structural properties.

## Mathematical Explanation
For any matrix A ∈ ℝᵐˣⁿ:

A = U Σ Vᵀ

Where:
- U is an orthogonal matrix (left singular vectors)
- Σ is a diagonal matrix of singular values
- Vᵀ is the transpose of an orthogonal matrix (right singular vectors)

Properties:
- Singular values are non-negative
- Rank equals number of non-zero singular values
- Works for non-square matrices
- More stable than eigen-decomposition

## Python Code
Compute SVD using NumPy.

```python
import numpy as np

A = np.array([[3, 1],
              [1, 3]])

U, S, Vt = np.linalg.svd(A)

U, S, Vt
```

## Matrix Reconstruction
Reconstruct A using U, Σ, and Vᵀ.

```python
Sigma = np.diag(S)

A_reconstructed = U @ Sigma @ Vt
A_reconstructed
```

## Low-Rank Approximation
Approximate A using top singular values.

```python
k = 1

U_k = U[:, :k]
S_k = np.diag(S[:k])
Vt_k = Vt[:k, :]

A_approx = U_k @ S_k @ Vt_k
A_approx
```

## Visualization
Singular values show importance of components.

```python
import matplotlib.pyplot as plt

plt.plot(S, marker='o')
plt.title("Singular Values")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid()
plt.show()
```
