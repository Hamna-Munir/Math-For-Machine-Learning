## Introduction

In Machine Learning, most functions depend on multiple variables.

Examples:
- Loss functions
- Cost functions
- Neural network outputs
- Optimization functions

These are called multivariable functions.


## Definition

A multivariable function takes more than one input variable.

Example:

f(x, y) = x² + y²  
f(x, y, z) = x + y + z


## Geometric Interpretation

A function f(x, y) produces a surface in 3D space.

z = f(x, y)

This helps visualize optimization problems.


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = sp.symbols('x y')

f = x**2 + y**2

f
```

## Function Evaluation

We can evaluate multivariable functions at specific points.


```python
f.subs({x: 2, y: 3})
```

## Partial Derivatives

To study how one variable affects the function, we use partial derivatives.

We differentiate with respect to one variable while keeping others constant.


```python
fx = sp.diff(f, x)
fy = sp.diff(f, y)

fx, fy
```

## Level Curves

Level curves show where the function has the same value.

They are useful in optimization.


```python
f_l = sp.lambdify((x, y), f, 'numpy')

x_vals = np.linspace(-5, 5, 200)
y_vals = np.linspace(-5, 5, 200)

X, Y = np.meshgrid(x_vals, y_vals)
Z = f_l(X, Y)

plt.contour(X, Y, Z, levels=20)
plt.colorbar()

plt.title("Level Curves of f(x, y) = x² + y²")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

## 3D Surface Visualization

3D plots help us understand the shape of functions.


```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, alpha=0.7)

ax.set_title("Surface: f(x, y) = x² + y²")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

plt.show()
```

## Machine Learning Connection

Multivariable functions appear in:

- Loss functions

- Deep learning layers
