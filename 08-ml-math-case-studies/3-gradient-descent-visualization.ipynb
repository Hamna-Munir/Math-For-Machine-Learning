## Introduction

Gradient Descent is an optimization algorithm used to minimize a function by iteratively moving in the direction of the steepest descent (negative gradient).

We'll use a **2D quadratic function** for easy visualization:

f(x, y) = x² + y²

The global minimum is at (0,0).

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Define Function and Gradient

f(x, y) = x² + y²  
Gradient: ∇f = [2x, 2y]

```python
def f(X):
    return X[0]**2 + X[1]**2

def grad_f(X):
    return np.array([2*X[0], 2*X[1]])
```

## Gradient Descent Function

```python
def gradient_descent(start, lr, iterations):
    path = [start]
    x = np.array(start, dtype=float)
    for i in range(iterations):
        grad = grad_f(x)
        x = x - lr * grad
        path.append(x.copy())
    return np.array(path)
```

## Run Gradient Descent

```python
start = [4, 4]
learning_rate = 0.1
iterations = 20

path = gradient_descent(start, learning_rate, iterations)
path
```

## 2D Contour Visualization

```python
# Create a grid for contour plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.figure(figsize=(8,6))
plt.contour(X, Y, Z, levels=30, cmap='viridis')
plt.plot(path[:,0], path[:,1], 'ro-', label='Gradient Descent Path')
plt.scatter([0], [0], color='blue', label='Minimum')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent on f(x, y) = x^2 + y^2')
plt.legend()
plt.show()
```

## 3D Surface Visualization

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.plot(path[:,0], path[:,1], np.array([f(p) for p in path]), 'r.-', label='GD Path')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Gradient Descent Path on 3D Surface')
plt.show()
```
