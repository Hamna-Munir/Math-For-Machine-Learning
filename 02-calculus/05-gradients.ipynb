## Concept

The gradient is a vector of partial derivatives that shows the direction of steepest increase of a function.

In Machine Learning, gradients guide how model parameters are updated.


## Mathematical Explanation

For a function f(x, y):

Gradient of f is:

∇f = [ ∂f/∂x , ∂f/∂y ]

If:

f(x, y) = x² + y²

Then:

∂f/∂x = 2x  
∂f/∂y = 2y

∇f = [2x , 2y]


## Python Implementation

We use SymPy to compute gradients symbolically.


```python
import sympy as sp

x, y = sp.symbols('x y')
```

```python
f = x**2 + y**2

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

df_dx, df_dy
```

```python
gradient = sp.Matrix([df_dx, df_dy])
gradient
```

## Geometric Interpretation

The gradient points in the direction of maximum increase.

Moving in the opposite direction of the gradient decreases the function.
This idea is used in Gradient Descent.


## Visualization

We visualize gradient directions on a contour plot.


```python
import numpy as np
import matplotlib.pyplot as plt

f_l = sp.lambdify((x, y), f, 'numpy')

x_vals = np.linspace(-4, 4, 100)
y_vals = np.linspace(-4, 4, 100)

X, Y = np.meshgrid(x_vals, y_vals)
Z = f_l(X, Y)

dx = 2 * X
dy = 2 * Y

plt.contour(X, Y, Z, levels=20)
plt.quiver(X, Y, dx, dy)

plt.title("Gradient Field of f(x,y) = x² + y²")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

## ML Connection

Gradients are used in:

- Gradient Descent

- Neural Network Training
