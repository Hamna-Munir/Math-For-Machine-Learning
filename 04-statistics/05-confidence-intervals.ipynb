## Introduction

A **confidence interval (CI)** provides a range of values that is likely to contain the true population parameter.

Instead of giving a single estimate, it gives a range with a certain level of confidence.

Example:

- We are 95% confident that the true mean lies within a specific interval.

## Key Concepts

- **Confidence Level:** Probability that the interval contains the true parameter (e.g., 95%)
- **Margin of Error:** Range above and below the sample estimate
- **Critical Value:** Based on distribution (z or t)


```python
import numpy as np
from scipy import stats
```

## Generate Sample Data

We generate a dataset from a normal distribution.

```python
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

mean
```

## 1. Confidence Interval using t-distribution

When population standard deviation is unknown, we use the **t-distribution**.

```python
confidence_level = 0.95

t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

margin_of_error = t_critical * (std / np.sqrt(n))

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print("Confidence Interval:", (lower_bound, upper_bound))
```

## 2. Using Built-in Function

We can also compute the confidence interval using SciPy.

```python
ci = stats.t.interval(confidence_level, df=n-1, loc=mean, scale=std/np.sqrt(n))
print("Confidence Interval (SciPy):", ci)
```

## Visualization

Let's visualize the confidence interval.

```python
import matplotlib.pyplot as plt

plt.hist(data, bins=30)
plt.axvline(lower_bound)
plt.axvline(upper_bound)
plt.axvline(mean)
plt.title("Confidence Interval Visualization")
plt.show()
```

## Interpretation

A 95% confidence interval means:

- If we repeat the experiment many times,
- About 95% of the intervals will contain the true population mean.


## Why Confidence Intervals Matter in Machine Learning

Confidence intervals are useful in Machine Learning for:

- Estimating model performance
- Understanding uncertainty in predictions
- Evaluating metrics (accuracy, mean error)
- Comparing models reliably
