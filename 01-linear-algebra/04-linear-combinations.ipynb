## Concept
A linear combination is formed by multiplying vectors by scalars and then adding the results together. Linear combinations describe how vectors can be combined to create new vectors.

## Mathematical Explanation
Given vectors v₁, v₂, ..., vₙ and scalars a₁, a₂, ..., aₙ, a linear combination is defined as:

a₁v₁ + a₂v₂ + ... + aₙvₙ

Linear combinations are used to define span, vector spaces, and model relationships between features.

## Python Code
Below is a simple example of computing a linear combination using NumPy.

```python
import numpy as np

v1 = np.array([1, 2])
v2 = np.array([3, 4])

a = 2
b = 1

a * v1 + b * v2
```

## Visualization
The plot below shows vectors and their linear combination in 2D space.

```python
import matplotlib.pyplot as plt

result = a * v1 + b * v2

plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, label='v2')
plt.quiver(0, 0, result[0], result[1], angles='xy', scale_units='xy', scale=1, label='Linear Combination')

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid()
plt.legend()
plt.title('Linear Combination of Vectors')
plt.show()
```

## ML Connection
In Machine Learning, predictions are often linear combinations of input features and weights. Linear regression, logistic regression, and neural networks all rely on weighted sums of inputs.

## Summary
- Linear combinations combine vectors using scalar weights
- They define span and vector spaces
- Machine Learning models compute weighted linear combinations of features
