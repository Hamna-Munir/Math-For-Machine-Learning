## Introduction

Probability distributions describe how the values of a random variable are distributed.

Several distributions appear frequently in **Machine Learning, statistics, and data science**.

In this notebook we will explore:

- Bernoulli Distribution
- Binomial Distribution
- Normal (Gaussian) Distribution
- Poisson Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Bernoulli Distribution

The Bernoulli distribution models an experiment with **two possible outcomes**.

Examples:

- Coin toss (Heads or Tails)
- Email is spam or not spam
- Student passes or fails

It has a single parameter **p**, the probability of success.

```python
p = 0.6
samples = np.random.binomial(1, p, 1000)

plt.hist(samples, bins=2)
plt.title("Bernoulli Distribution")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()
```

## 2. Binomial Distribution

The Binomial distribution models the number of successes in **multiple independent Bernoulli trials**.

Example:

- Number of heads in 10 coin tosses
- Number of successful predictions in a set of trials

```python
n = 10
p = 0.5

binomial_samples = np.random.binomial(n, p, 1000)

plt.hist(binomial_samples, bins=10)
plt.title("Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

## 3. Normal (Gaussian) Distribution

The Normal distribution is one of the most important distributions in statistics and Machine Learning.

It is characterized by:

- Mean (μ)
- Standard deviation (σ)

Many real-world phenomena follow a normal distribution such as height, measurement errors, and noise in data.

```python
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30)
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## 4. Poisson Distribution

The Poisson distribution models the number of events occurring in a fixed interval of time or space.

Examples:

- Number of website visits per minute
- Number of calls received in a call center
- Number of defects in manufacturing

```python
lam = 4
poisson_samples = np.random.poisson(lam, 1000)

plt.hist(poisson_samples, bins=15)
plt.title("Poisson Distribution")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.show()
```

## Why These Distributions Matter in Machine Learning

These distributions appear frequently in Machine Learning models:

- **Bernoulli & Binomial** → classification problems
- **Normal distribution** → Gaussian models and noise assumptions
- **Poisson distribution** → modeling count data

Many ML algorithms assume that data follows specific probability distributions.
