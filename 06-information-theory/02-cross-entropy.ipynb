## Introduction

Cross-Entropy measures the difference between two probability distributions:

- True distribution (actual labels)
- Predicted distribution (model output)

It is widely used as a **loss function in classification problems**.

## Cross-Entropy Formula

H(p, q) = - Σ p(x) log₂ q(x)

Where:
- p(x) → True distribution
- q(x) → Predicted distribution


## Intuition

- If predictions are accurate → Low cross-entropy
- If predictions are wrong → High cross-entropy

The goal of training is to **minimize cross-entropy**.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Cross-Entropy Implementation

```python
def cross_entropy(p, q):
    p = np.array(p)
    q = np.array(q)
    q = np.clip(q, 1e-10, 1.0)  # avoid log(0)
    return -np.sum(p * np.log2(q))
```

## Example

```python
# True distribution (one-hot)
p = [1, 0, 0]

# Good prediction
q_good = [0.9, 0.05, 0.05]

# Bad prediction
q_bad = [0.2, 0.4, 0.4]

print("Cross-Entropy (Good Prediction):", cross_entropy(p, q_good))
print("Cross-Entropy (Bad Prediction):", cross_entropy(p, q_bad))
```

## Visualization

```python
q = np.linspace(0.01, 0.99, 100)
loss = [-np.log2(x) for x in q]

plt.plot(q, loss)
plt.title("Cross-Entropy Loss for Binary Classification")
plt.xlabel("Predicted Probability (Correct Class)")
plt.ylabel("Loss")
plt.show()
```

## Relationship with Entropy

Cross-Entropy = Entropy + KL Divergence

H(p, q) = H(p) + D_KL(p || q)


## Why Cross-Entropy is Important

- Used as loss function in classification models
- Works well with probabilistic outputs
- Helps optimize model predictions


## Applications in Machine Learning

- Logistic Regression
- Neural Networks
- Deep Learning classification tasks
