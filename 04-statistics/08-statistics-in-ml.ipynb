## Introduction

Statistics plays a fundamental role in Machine Learning.

It helps us:

- Understand data
- Make predictions
- Evaluate models
- Handle uncertainty

Almost every ML algorithm is built on statistical principles.

## Key Statistical Concepts Used in ML

- Descriptive Statistics (mean, median, variance)
- Probability Distributions
- Hypothesis Testing
- Confidence Intervals
- Correlation and Covariance
- Bias-Variance Tradeoff


## Example: Data Understanding

```python
import numpy as np

data = np.random.normal(50, 10, 1000)

print("Mean:", np.mean(data))
print("Variance:", np.var(data))
```

## Example: Feature Relationships

```python
x = np.random.rand(100)
y = 3 * x + np.random.normal(0, 0.2, 100)

correlation = np.corrcoef(x, y)[0, 1]
print("Correlation:", correlation)
```

## Example: Model Evaluation Concept

We use statistics to evaluate model performance using metrics like:

- Accuracy
- Precision & Recall
- Mean Squared Error (MSE)
- R-squared


## Example: Handling Uncertainty

Confidence intervals and probability help quantify uncertainty in predictions.

This is important for making reliable decisions in real-world applications.

## Visualization

```python
import matplotlib.pyplot as plt

plt.hist(data, bins=30)
plt.title("Data Distribution")
plt.show()
```

## Real ML Applications

- Linear Regression → Assumes statistical relationships
- Logistic Regression → Uses probability
- Naive Bayes → Based on Bayes theorem
- A/B Testing → Hypothesis testing
- Regularization → Bias-variance tradeoff
