## Introduction

In optimization, the shape of the function determines how easy it is to find the minimum.

- Convex functions → Easy to optimize
- Non-convex functions → Harder to optimize

This distinction is crucial in Machine Learning.

## Convex Functions

A function is convex if:

- It has a single global minimum
- No local minima
- The curve is bowl-shaped

Example: f(x) = x²

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y_convex = x**2

plt.plot(x, y_convex)
plt.title("Convex Function")
plt.show()
```

## Non-Convex Functions

A function is non-convex if:

- It has multiple local minima and maxima
- Optimization algorithms can get stuck

Example: f(x) = x³ - 3x

```python
y_nonconvex = x**3 - 3*x

plt.plot(x, y_nonconvex)
plt.title("Non-Convex Function")
plt.show()
```

## Key Differences

| Property | Convex | Non-Convex |
|----------|--------|------------|
| Minima | One global minimum | Multiple minima |
| Optimization | Easy | Difficult |
| Stability | High | Lower |


## Why It Matters in Machine Learning

- Convex problems guarantee optimal solutions
- Non-convex problems (like deep learning) are harder to solve
- Optimization algorithms must handle complex landscapes


## Real ML Insight

- Linear Regression → Convex problem
- Neural Networks → Non-convex problem

This is why training deep learning models is challenging.
