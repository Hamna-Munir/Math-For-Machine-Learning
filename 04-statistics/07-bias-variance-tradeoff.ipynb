## Introduction

In Machine Learning, models can make two types of errors:

- **Bias**: Error due to overly simple assumptions
- **Variance**: Error due to sensitivity to small fluctuations in data

The goal is to find a balance between them to achieve good generalization.

## Understanding Bias

- High bias → Model is too simple
- Leads to **underfitting**
- Cannot capture patterns in data


## Understanding Variance

- High variance → Model is too complex
- Leads to **overfitting**
- Captures noise along with patterns


## Visualization Intuition

- High Bias → Straight line for complex data
- High Variance → Very wiggly curve
- Ideal Model → Smooth curve capturing real pattern


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

## Generate Sample Data

```python
np.random.seed(42)
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = np.sin(X).ravel() + np.random.normal(0, 0.3, 100)
```

## Fit Models with Different Complexity

```python
def plot_model(degree):
    poly = PolynomialFeatures(degree)
    X_poly = poly.fit_transform(X)
    model = LinearRegression()
    model.fit(X_poly, y)
    
    y_pred = model.predict(X_poly)
    
    plt.scatter(X, y)
    plt.plot(X, y_pred)
    plt.title(f"Degree {degree}")
    plt.show()

plot_model(1)   # High Bias
plot_model(5)   # Balanced
plot_model(15)  # High Variance
```

## Bias-Variance Tradeoff

- Increasing model complexity:
  - Bias ↓
  - Variance ↑

- Decreasing model complexity:
  - Bias ↑
  - Variance ↓


## Error Decomposition

Total Error = Bias² + Variance + Irreducible Error


## Why It Matters in Machine Learning

- Helps in choosing the right model complexity
- Guides regularization techniques
- Important for model evaluation
- Improves generalization performance
