## Introduction

Matrix conditioning tells us how sensitive the solution of a system is to small changes in input.

- Well-conditioned → Stable results
- Ill-conditioned → Unstable results


## Condition Number

Condition Number (κ) measures sensitivity:

κ(A) = ||A|| × ||A⁻¹||

- Small κ → Stable
- Large κ → Unstable


```python
import numpy as np
```

## Example: Well-Conditioned Matrix

```python
A = np.array([[2, 0],
              [0, 2]])

cond_A = np.linalg.cond(A)
print("Condition Number (A):", cond_A)
```

## Example: Ill-Conditioned Matrix

```python
B = np.array([[1, 1],
              [1, 1.0001]])

cond_B = np.linalg.cond(B)
print("Condition Number (B):", cond_B)
```

## Solving Linear Systems

Solve Ax = b and observe sensitivity.

```python
b = np.array([2, 2])

x_A = np.linalg.solve(A, b)
x_B = np.linalg.solve(B, b)

print("Solution (A):", x_A)
print("Solution (B):", x_B)
```

## Effect of Small Changes

```python
b_perturbed = np.array([2, 2.0001])

x_B_perturbed = np.linalg.solve(B, b_perturbed)

print("Original Solution (B):", x_B)
print("Perturbed Solution (B):", x_B_perturbed)
```

## Key Insights

- Ill-conditioned matrices amplify errors
- Small input changes → Large output changes
- High condition number is dangerous


## Applications in Machine Learning

- Linear Regression (normal equation)
- Neural network training
- Optimization problems
- Numerical stability analysis


## How to Handle Ill-Conditioning

- Feature scaling
- Regularization (Ridge)
- Using SVD instead of matrix inversion
