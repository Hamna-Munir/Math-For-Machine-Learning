## Introduction

Probability is one of the most important mathematical foundations of **Machine Learning**.

Machine Learning models often work with **uncertainty**, incomplete data, and randomness. Probability provides the tools to handle these situations.

Many popular algorithms are based directly on probabilistic principles.

## Why Probability Matters in Machine Learning

Probability helps Machine Learning systems:

- Handle uncertainty in data
- Model randomness and noise
- Estimate likelihood of outcomes
- Make predictions based on observed data
- Learn patterns from data distributions

## 1. Naive Bayes Classifier

Naive Bayes is a popular Machine Learning algorithm based on **Bayes' Theorem**.

It is commonly used for:

- Spam detection
- Text classification
- Sentiment analysis

The algorithm assumes that features are conditionally independent given the class label.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X, y = make_classification(n_samples=1000, n_features=5, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = GaussianNB()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print("Model Accuracy:", accuracy)
```

## 2. Probability Distributions in ML

Many Machine Learning models assume that data follows certain probability distributions.

Examples:

- **Normal distribution** → used in Gaussian models
- **Bernoulli distribution** → used in binary classification
- **Poisson distribution** → used for count data

Understanding these distributions helps in choosing the right model.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0,1,1000)

plt.hist(data, bins=30)
plt.title("Example of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## 3. Likelihood and Model Training

Many Machine Learning algorithms learn parameters by **maximizing likelihood**.

The idea is simple:

Find parameters that make the observed data most probable.

This concept is called **Maximum Likelihood Estimation (MLE)**.

## 4. Uncertainty in Predictions

Probability allows Machine Learning models to output **confidence levels** instead of just predictions.

For example:

- Email spam probability = 0.92
- Image contains a cat = 0.87

These probabilities help in decision making.
