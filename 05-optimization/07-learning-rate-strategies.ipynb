## Introduction

The learning rate is one of the most important hyperparameters in Machine Learning.

- Too high → Divergence
- Too low → Slow learning

Learning rate strategies help adjust it dynamically during training.

## Types of Learning Rate Strategies

1. Constant Learning Rate
2. Time-Based Decay
3. Step Decay
4. Exponential Decay
5. Learning Rate Scheduling


```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Constant Learning Rate

```python
def constant_lr(initial_lr, epoch):
    return initial_lr
```

## 2. Time-Based Decay

lr = lr / (1 + decay * epoch)

```python
def time_decay_lr(initial_lr, decay, epoch):
    return initial_lr / (1 + decay * epoch)
```

## 3. Step Decay

Reduce learning rate at fixed intervals.

```python
def step_decay_lr(initial_lr, drop, epochs_drop, epoch):
    return initial_lr * (drop ** (epoch // epochs_drop))
```

## 4. Exponential Decay

lr = lr * exp(-decay * epoch)

```python
def exponential_decay_lr(initial_lr, decay, epoch):
    return initial_lr * np.exp(-decay * epoch)
```

## Visualization of Strategies

```python
epochs = np.arange(0, 50)

constant = [constant_lr(0.1, e) for e in epochs]
time_decay = [time_decay_lr(0.1, 0.05, e) for e in epochs]
step_decay = [step_decay_lr(0.1, 0.5, 10, e) for e in epochs]
exp_decay = [exponential_decay_lr(0.1, 0.05, e) for e in epochs]

plt.plot(epochs, constant, label='Constant')
plt.plot(epochs, time_decay, label='Time Decay')
plt.plot(epochs, step_decay, label='Step Decay')
plt.plot(epochs, exp_decay, label='Exponential Decay')

plt.legend()
plt.title("Learning Rate Strategies")
plt.xlabel("Epochs")
plt.ylabel("Learning Rate")
plt.show()
```

## Why Learning Rate Strategies Matter

- Improve convergence speed
- Prevent overshooting
- Help reach better minima
