## What is Optimization?

Optimization is the process of finding the **best possible solution** to a problem.

In Machine Learning, optimization means:

- Minimizing error (loss)
- Improving model predictions
- Finding optimal parameters (weights)


## Real-Life Intuition

Imagine you are standing on a hill and want to reach the lowest point (valley).

- You look at the slope
- Move step by step downward
- Eventually reach the minimum point

This is exactly how optimization algorithms like Gradient Descent work.

## Objective Function

An objective function (or loss function) measures how wrong a model is.

Goal:

- Minimize the loss function

Example:

- Mean Squared Error (MSE)
- Cross-Entropy Loss


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Simple Example Function

We define a simple quadratic function:

f(x) = x²

The goal is to find the minimum value.

```python
x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y)
plt.title("Objective Function f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## Minimum Point

From the graph, we can see:

- The minimum occurs at x = 0
- f(x) = 0 is the lowest value


## Why Optimization is Important in ML

Optimization is used in:

- Training models (finding best weights)
- Reducing prediction error
- Improving model accuracy
- Deep Learning (Neural Networks)


## Types of Optimization Problems

- Minimization (most common in ML)
- Maximization (e.g., likelihood functions)


## Key Takeaways

- Optimization helps find the best model parameters
- It works by minimizing a loss function
- It is the backbone of Machine Learning algorithms
